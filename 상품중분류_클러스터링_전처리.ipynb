#대분류 상대

```python
dl=uniq_df.l_pd.value_counts()
lsum=sum(uniq_df.l_pd.value_counts())
```

```python
l_p=[]
for p in uniq_df.l_pd:
  l_p.append(round(dl[p]/lsum,3))
l_p[:3]
```

#중분류 상대

```python
dm=uniq_df.m_pd.value_counts()
msum=sum(uniq_df.m_pd.value_counts())
```

```python
m_p=[]
for p in uniq_df.m_pd:
  m_p.append(round(dm[p]/msum,3))
```

#동일 대분류 내 중분류 중 상대

```python
l_m_p=[]
for x,y in zip(uniq_df.l_pd,uniq_df.m_pd):
  l_m_p.append(round(dm[y]/dl[x],3))
```

```python
for x,y,z in zip(l_p[:10],m_p[:10],l_m_p[:10]):
  print(x,y,z)

#중분류 종류가 많아서 퍼센트가 낮게 잡히고, 대분류내 중분류가 높게 잡힘
```

```python
uniq_df['l_p']=l_p
uniq_df['m_p']=m_p
uniq_df['l_m_p']=l_m_p
```

#성별&연령

```python
def sex_(x):
  if x=='여성':
    return 1
  else:
    return 0
```

```python
def age_(x):
    return int(x[:-1])+5
```

```python
uniq_df['ma_fem_dv']=uniq_df['ma_fem_dv'].apply(sex_)
```

```python
uniq_df['ages']=uniq_df['ages'].apply(age_)
```

```python
uniq_df.head()
```

```python
sex=[]
age=[]

for p in mp:
  sex.append(uniq_df[uniq_df['m_pd']==p].ma_fem_dv.mean().round(3))
  age.append(uniq_df[uniq_df['m_pd']==p].ages.mean().round(3))
```
