```python
import re
f=open('/content/wth.txt','r',encoding='utf8')
wth=f.read()
```

```python
w=re.findall(r'<span>(.+)</span>', wth)
```

```python
w
```

```python
weather=[]
for s in w:
  if s[0]=='일' or s[0]=='평' or s[0]=='최' or s[-1]=='일':
    weather.append(s)
```

```python
date=[]
avgtemp=[]
avgrain=[]

#date

m=0
for wt in w:
  if wt[-1]=='일':
    if wt=='1일':
      m+=1
    date.append(str(m)+'.'+wt[:-1])

  if wt[0]=='평' and wt[-1]=='℃':
    avgtemp.append(float(wt[5:-1]))

  if wt[:2]=='일강':
    avgrain.append(wt[5:])
```
