```python
import io
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
demo = pd.read_csv("/content/drive/MyDrive/롯데 공모전/dataset/LPOINT_BIG_COMP_01_DEMO.csv") #고객 데모 정보
pdde = pd.read_csv("/content/drive/MyDrive/롯데 공모전/dataset/LPOINT_BIG_COMP_02_PDDE.csv") #상품 구매 정보(유통사 상품 구매 내역)
pd_clac = pd.read_csv("/content/drive/MyDrive/롯데 공모전/dataset/LPOINT_BIG_COMP_04_PD_CLAC.csv") #상품 분류 정보(유통사 상품 카테고리 마스터)
```

```python
df = pd.merge(pdde, pd_clac,how='outer',on='pd_c')
df = pd.merge(df, demo, how='left', on='cust')
df.head()
```

```python
from datetime import datetime

a = df['de_dt'].astype('str')
df['de_dt']  = pd.to_datetime(a)
df['day'] = df['de_dt'].map(lambda x : x.strftime('%A'))
df['month'] = df['de_dt'].dt.month
df['day'] = df['de_dt'].dt.day

df.head()
```

##### 식료품 추출

```python
eat = '채소|과일|과자|조리식품|조미료|수산물|축산물|건해산물|커피/차|냉동식품|냉장식품|대용식|병통조림|양곡|유제품|음료|주류'
eat_df = df[df['clac_hlv_nm'].str.contains(eat)]
eat_df = eat_df.reset_index()
eat_df.head()
```

- 고객의 총 주문 횟수, 고객이 총 주문한 상품 수, 고객이 총 주문한 상품 종류의 수

```python
order1 = eat_df.groupby('cust').sum()['buy_ct'].rename_axis('cust').reset_index(name='total_items') #고객의 총 주문한 상품 수
order1.head()
```

```python
order2 = eat_df[['rct_no']].value_counts().reset_index(name='one_times_items') #한번에 주문한 상품의 수
order2.head()
```

```python
order3 = eat_df.groupby('cust')['rct_no'].nunique().reset_index(name='rct_cnt') #구매 횟수
order3.head()
```

```python
df = pd.merge(eat_df, order1, how='outer',on='cust')
df = pd.merge(df, order2, how='outer',on='rct_no')
df = pd.merge(df, order3, how='outer',on='cust')

df.head()
```

- 재구매 여부

```python
rebuy_df = eat_df.groupby(['cust', 'clac_hlv_nm']).sum()['buy_ct'].reset_index(name='rebuy_cnt') 
rebuy_df['rebuy_cnt'] -= 1

rebuy_df['rebuy'] = 0
rebuy_df.loc[rebuy_df['rebuy_cnt'] >= 1,'rebuy'] = 1
rebuy_df = rebuy_df[['cust','clac_hlv_nm','rebuy']]
rebuy_df.head()
```

```python
df = pd.merge(df, rebuy_df, how='outer',on=['cust','clac_hlv_nm'])
df.head()
```

- 총 재구매 횟수

```python
rebuy_df = eat_df.groupby(['cust', 'clac_hlv_nm']).sum()['buy_ct'].reset_index(name='rebuy_cnt') 
rebuy_df['rebuy_cnt'] -= 1
rebuy_df = rebuy_df.groupby('cust').sum()['rebuy_cnt'].reset_index(name='rebuy_cnt') 

rebuy_df.head()
```

```python
df = pd.merge(df, rebuy_df, how='outer',on='cust')
df.head()
```

```python
(df['rebuy_cnt']/df['total_items']*100).min()
```

```python
df['rebuy_ration'] = df['rebuy_cnt']/df['total_items']*100
```

```python
df['total_am'] = df['buy_am'] * df['buy_ct']
```

```python
df_agg = df.groupby(['cust', 'rct_no'])[['total_am','buy_ct','buy_am']].agg(['sum','mean','std'])
df_agg.columns = list(map(lambda x:x[0]+'_'+x[1],list(df_agg)))
```

```python
df = df.sort_values(['cust','de_dt'])
df['de_diff'] = df['de_dt'].diff()
df.loc[df['cust'] != df['cust'].shift(),'de_diff'] = None
```
